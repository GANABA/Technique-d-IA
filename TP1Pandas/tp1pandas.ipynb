```python
# Import 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
# Chargement de la dataset
iris = load_iris()
```

```python
iris
```

```python
# Transformation en dataframe
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
# Ajouter la colonne target (valeurs numériques 0, 1, 2)
iris_df['target'] = iris.target

# Ajouter une colonne avec le nom de l'espèce (setosa, versicolor, virginica)
iris_df['Especes'] = iris.target_names[iris.target]
```

```python
# Affichage des cinq première lignes
iris_df.head()
```

```python
# Taille de la dataset
iris_df.shape
```

```python
# Statistiques descriptives
iris_df.describe()
```

```python
iris_df.info()
```

```python
# Renommer les colonnes
new_iris_df = iris_df.rename(columns = {
    'sepal length (cm)' : 'SepalLengthCm',
    'sepal width (cm)' : 'SepalWidthCm',
    'petal length (cm)' : 'PetalLengthCm',
    'petal width (cm)' : 'PetalWidhtCm'
})
```

```python
print(new_iris_df.columns)
```

```python
# Calcul du rapport longueur sepal/largeur sepal dans une nouvelle colonne
new_iris_df['RapportLongLargPetal'] = new_iris_df['PetalLengthCm'] / new_iris_df['PetalWidhtCm']
```

```python
# Nouvelle dataset
new_iris_df.head()
```

```python
# Suppression d'une colonne
new_iris_df.drop('SepalWidthCm', axis=1, inplace=True)
```

```python
# Nouvelle dataset après suppression d'1 colonne
new_iris_df.head()
```

```python
# Supprimer toutes les lignes où la longueur du sépale est inférieure à 5.0
new_iris_df.drop(new_iris_df.loc[new_iris_df["SepalLengthCm"] < 5.0 ].index, inplace=True)
```

```python
# Nouvelle dataset après suppression de ligne sous condition
new_iris_df.head()
```

```python
# Taille de la dataset après suppresion de ligne
new_iris_df.shape
```

```python
# Filtrer uniquement les lignes correspondant à l’espèce 'setosa'
setosa_df = new_iris_df.loc[new_iris_df['target'] == 0]
```

```python
# Compter le nombre d’occurrences de chaque espèce
new_iris_df['Especes'].value_counts()
```

```python
# Visualiser : (1) un histogramme d’une variable
new_iris_df['PetalLengthCm'].hist()
```

```python
# Visualiser : (2) un nuage de points entre deux variables
plt.scatter(new_iris_df['PetalLengthCm'], new_iris_df['PetalWidhtCm'])
plt.title("Nuage de points : Longueur vs Largeur des pétales")
plt.xlabel("Longueur du pétale (cm)")
plt.ylabel("Largeur du pétale (cm)")
plt.show()
```

```python
# Visualiser :  (3) un boxplot comparant les espèces
sns.boxplot(x='Especes', y='PetalLengthCm', data=new_iris_df)
plt.title("Répartition de la longueur des pétales selon l'espèce")
plt.xlabel("Espèce")
plt.ylabel("Longueur du pétale (cm)")
plt.show()
```
